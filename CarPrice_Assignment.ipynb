```python
plt.pie(df.car_ID, labels = df.doornumber, autopct='%1.2f%%')
plt.title('Cars')
plt.legend(df.doornumber, loc='lower right',title="Distribution:")
plt.show()

```

```python
df
```

```python
#using bar graph
plt.bar(df.carbody,df.doornumber)
plt.title('Cars')
plt.xlabel('carbody')
plt.ylabel('Door Number')
plt.legend(df.doornumber, loc='lower right',title="Distribution:")
plt.show()
```

```python
#to convert the above vertical bar graph to horizontal
plt.barh(df.car_ID,df.doornumber)
plt.title('Cars')
plt.xlabel('Car ID')
plt.ylabel('Door Number')
plt.legend(df.doornumber, loc='lower right',title="Distribution:")
plt.show()
```

```python
#using stack bar
plt.bar(df.car_ID,df.doornumber)
plt.show
```

```python
df
```

```python
#stacked bar graph
plt.bar(df.car_ID,df.fueltype,color='r')
plt.bar(df.car_ID,df.doornumber, bottom=df.fueltype, color='b')
plt.show()
```

```python
# Data
x = ['A', 'B', 'C', 'D']
y1 = [10, 20, 0, 30]
y2 = [20, 25, 15, 25]

# Plotting
plt.bar(x, y1, color='r')
plt.bar(x, y2, bottom=y1, color='b')
plt.show()
```
