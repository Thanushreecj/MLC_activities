DATA CLEANING QUESTIONS – RAW_DATASET.CSV

|                                 |
|:--------------------------------|
| SECTION A: COLUMN NAME CLEANING |

1.  Identify column names containing leading/trailing spaces.
2.  Detect inconsistent capitalization in column headers.
3.  Rename ‘irst Name’ to ‘First Name’.
4.  Convert all column names to snake_case format.

|                                 |
|:--------------------------------|
| SECTION B: DATA TYPE VALIDATION |

1.  Print data types of all columns.
2.  Identify columns incorrectly typed as object but should be numeric
    or datetime.
3.  Convert the ‘Age’ column (which contains values like 10/28/2019) to
    datetime.
4.  Assuming ‘Age’ represents date of birth, compute age_years.
5.  Identify invalid date entries and handle them.

|                            |
|:---------------------------|
| SECTION C: SALARY CLEANING |

1.  Convert ‘SALARY’ from currency strings (e.g., \$92.32) to float.
2.  Identify rows where salary conversion fails.
3.  Detect missing or malformed salary values.
4.  Compute minimum, maximum, and average salary after cleaning.
5.  Flag salaries equal to zero or negative.

|                          |
|:-------------------------|
| SECTION D: NAME CLEANING |

1.  Remove leading/trailing spaces from names.
2.  Convert names to proper title case.
3.  Detect names containing digits or special characters.
4.  Create a full_name column combining First Name and Last name.

|                                        |
|:---------------------------------------|
| SECTION E: EMAIL CLEANING & VALIDATION |

1.  Convert all emails to lowercase.
2.  Remove leading/trailing spaces in email column.
3.  Identify missing email addresses.
4.  Detect duplicate emails.
5.  Validate email format using regex.
6.  Extract email domain and compute frequency of each domain.

|                             |
|:----------------------------|
| SECTION F: ADDRESS CLEANING |

1.  Identify addresses containing ‘P.O. Box’.
2.  Detect inconsistent abbreviations (Rd. vs Road, St. vs Street).
3.  Concatenate Address1, Address2, and Address3 into full_address.
4.  Detect duplicate addresses.

|                                        |
|:---------------------------------------|
| SECTION G: MISSING VALUES & DUPLICATES |

1.  Compute total missing values per column.
2.  Calculate percentage of missing values.
3.  Remove fully duplicate rows.
4.  Identify soft duplicates (same email but different name).

|                                    |
|:-----------------------------------|
| SECTION H: DATA CONSISTENCY CHECKS |

1.  Check for unrealistic ages (negative or greater than 100).
2.  Detect salary outliers using IQR method.
3.  Determine whether email can serve as a primary key.

|                               |
|:------------------------------|
| SECTION I: EXPORT & REPORTING |

1.  Export cleaned dataset as cleaned_dataset.csv.
2.  Create a data quality report summarizing:
    -   Rows before cleaning
    -   Rows after cleaning
    -   Duplicates removed
    -   Invalid emails detected
    -   Salary conversion issues
    -   Date parsing issues
